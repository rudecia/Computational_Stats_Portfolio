## New Job Details

Enter the details of the new job below, as well as the google drive folder where you would like it to be saved:

```python
new_job_title = "Data Scientist" # @param {"type":"string"}
new_company = "Google" # @param {"type":"string"}
special_quality = "I like that you guys have free breakfast. That's a nice perk right there." # @param {"type":"string"}
folder_name = "Application Materials" # @param {"type":"string"}
file_path = f'drive/MyDrive/{folder_name}/'



!pip install python-docx
#run the above line if it is not already installed


from docx import Document

from google.colab import drive
drive.mount('/content/drive')

import re

if str.lower(old_job_title[:1]) in ['a', 'e', 'i', 'o', 'u']:
  old_job_title = 'an ' + old_job_title
else:
  old_job_title = 'a ' + old_job_title




def cover_letter(company: str, role: str, file_path: str, special_quality = ''):


    p0 = greeting

    #cleanly replace job titles by modifying their associated articles:
    if str.lower(role[:1]) in ['a', 'e', 'i', 'o', 'u']:
        role = 'an ' + role
    else:
        role = 'a ' + role


    if special_quality == '':
       lines = [p0]
       for p in paragraphs:
        if p != '':
            lines.append(p)

       lines.append(special_quality)
       para5 = f'{salutation[0]}, \n{salutation[1].lstrip()}'
       lines.append(para5)

    else:
      lines = [p0]
      for p in paragraphs:
        if p != '':
            lines.append(p)

      para5 = special_quality
      para6 = f'{salutation[0]}, \n{salutation[1].lstrip()}'

      lines.append(para5)
      lines.append(para6)



    #Save the file to desired folder
    file_name = file_path + company + ' Cover Letter.docx'

    doc = Document()

    for line in lines:
      line = re.sub(old_job_title, role, line, flags = re.IGNORECASE)
      line = re.sub(old_company, company, line, flags = re.IGNORECASE)



      doc.add_paragraph(line)

    doc.save(file_name)



cover_letter(new_company, new_job_title, file_path, special_quality)
```
