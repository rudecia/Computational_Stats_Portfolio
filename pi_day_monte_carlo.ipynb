plot

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10**5

def circle(x):
  return np.sqrt(1- x**2)


vals = np.random.uniform(size = (N, 2))

inside = []
outside = []

for x, y in vals:
  if y <= circle(x):
    inside.append((x,y))
  else:
    outside.append((x,y))


inside = np.array(inside)
outside = np.array(outside)


plt.scatter(inside[:, 0],inside[:, 1], c = 'g')
plt.scatter(outside[:, 0], outside[:, 1], c = 'r')
plt.title('Monte Carlo Estimate of Pi (Area Method)')
plt.show()




```

distribution of the estimates

```python

trials = 500

pi_estimates = np.zeros(trials)

for i in range(trials):

  x = np.random.uniform(size = N)
  y = np.random.uniform(size = N)

  #estimate of pi
  est = np.mean(y <= circle(x)) * 4

  pi_estimates[i] = est






plt.hist(pi_estimates)
plt.show()
```

mean of estimates

```python
np.mean(pi_estimates)
```
